```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

def plot_frequency_matrix(freq_matrix, title, alpha, k):
    # Define plot aesthetics
    plt.figure(figsize=(8, 6))
    sns.set(style="whitegrid")  # Clean, minimalistic style
    
    # Normalize frequency matrix to probabilities
    freq_matrix_prob = freq_matrix / np.sum(freq_matrix)
    
    # Create a custom light color palette
    cmap = sns.light_palette("skyblue", as_cmap=True)
    
    # Plot heatmap with rounded values
    sns.heatmap(
        freq_matrix_prob, 
        annot=True, 
        fmt=".2f", 
        cmap=cmap, 
        linewidths=0.5, 
        linecolor="white", 
        square=True, 
        cbar_kws={"shrink": 0.8},
        annot_kws={"fontsize": 12, "fontweight": "bold"}
    )
    
    # Customize axis labels
    plt.xticks([0.5, 1.5], ["Action 0 (P1)", "Action 1 (P1)"], fontsize=12)
    plt.yticks([0.5, 1.5], ["Action 0 (P0)", "Action 1 (P0)"], fontsize=12)
    
    # Add title with alpha and k values
    plt.title(f"{title}\nα={alpha}, k={k}", fontsize=14, fontweight='bold', pad=20)
    
    # Remove borders for a cleaner look
    plt.gca().spines['top'].set_visible(False)
    plt.gca().spines['right'].set_visible(False)
    
    plt.tight_layout()
    plt.show()


# Example Usage:
# Simulate or load a sample frequency matrix (2x2 for binary actions)
sample_freq_matrix = np.array([[30, 10], [5, 55]])
alpha = 0.3  # Example parameter
k = 0.005    # Example parameter

# Plot the frequency matrix
plot_frequency_matrix(sample_freq_matrix, "Frequency Matrix of Action Profiles", alpha, k)
```
